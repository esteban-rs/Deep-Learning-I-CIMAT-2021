El modelo de redes profundas Redes Antagónicas Generadoras (Generative Adversarial Networks, GANs) fué propuesto por Goodfellow et al., 2014b. Las GANs son un modelo generador (de los muchos que existen) que se basa en redes neuronales profundas.


En esta sección reformularemos el modelo de la GAN basadas en Perceptrones Multicapa a Convolucionales (Radford and Metz, 2015). La siguiente figura ilustra (a manera de recordatorio) la arquitectura de una red convolucional. Detalles de implementación en Keras pueden encontrarse en Dumoulin and Visin (2016), notas convnets 1, notes convnets 2 y Chollet (2018)

El modelo general seguirá siendo el de una GAN: un generador y un discriminador. El propósito del Generador $ G $ es generar datos sintéticos $ \tilde{x} $ que sean indistinguibles de los datos reales para el Discriminador $ D $. Por lo que el discriminador es una red que clasifica datos que se le presentan entre reales o falsos (fakes). La siguiente figura representa una GAN.

gan.png

Función Objetivo de la WGAN

Sean $ y \in \{0,1\} $ la etiqueta del dato $ x $ correspondientes a “falso” o “real” y $ \hat{y} \in [0,1] $ su estimación obtenida del discriminador para el dato en cuestión. Luego, la función a optimizar mediante el entrenamiento de la WGAN esta dada por

$$ L = \dfrac{1}{n} || W (v, \hat{v}) ||^2 $$

con

$$ W(v, \hat{v} ) = \dfrac{1}{\# B} \sum_{i \in B} v_{ij} - \dfrac{1}{\# \hat{B}} \sum_{i \in \hat{B}} \hat{v}_{ij}$$

donde $B$ define al conjunto de imágenes reales y $\hat{B} $ al conjunto de imágenes generadas.

```python
def reset_weights(model):
    '''
    Reinicializa los pesos del modelo 
    '''
    for ix, layer in enumerate(model.layers):
        if hasattr(model.layers[ix], 'kernel_initializer') and \
                hasattr(model.layers[ix], 'bias_initializer'):
            weight_initializer = model.layers[ix].kernel_initializer
            bias_initializer   = model.layers[ix].bias_initializer

            old_weights, old_biases = model.layers[ix].get_weights()

            model.layers[ix].set_weights([
                weight_initializer(shape=old_weights.shape),
                bias_initializer  (shape  =len(old_biases))])
```

# Generador

En generador es una red convolucional que

1. Toma como entrada un vector aleatorio de 20 entradas.

2. El cuál es pasado por una capa densa de dimensiones $(7)(7)(256)$ y luego reformateado (reshape) a un tensor de forma $ 7 \times 7 \times 256 $.

3. Luego es pasado a una convolución transpuesta que convierte el tensor $ 7 \times 7 \times 256 $ en uno de $ 14 \times 14 \times 128 $. Con posterior normalización or lotes y actrivación ReLU.

3. Luego es pasado a una convolución transpuesta que convierte el tensor $ 14 \times 14 \times 128 $ en uno de $ 14 \times 14 \times 64$. Con posterior normalización or lotes y actrivación ReLU.

4. Y de nuevo usando una convolución transpuesta se transforma el tensor $ 14 \times 14 \times 64 $ en uno del tamaño de la imagen de salida $ 28 \times 28 \times 1 $ a la que se la aplica una activación tanh.

Esto se ilustra en la siguiente figura

dcgan_generador.png

```python
from tensorflow.keras.utils import plot_model

def build_generator(img_shape, z_dim, verbose=False):
    '''
    Genera una imagen de 28x28x1 a partir de un vector aleatorio de 100 entradas (espacio latente)
    '''
    # Batch de datos
    z           = Input(shape=(z_dim,))
    # Pasa entrada unidimensional de dimensión 20 en un tensor de (7)(7)(256) tensor via un red Densa
    # luego la reformatea en un tensor de 7x7x128
    X = Dense(256 * 8 * 8, input_dim=z_dim) (z)
    X = Reshape((8, 8, 256))(X)

    # Convolución transpuesta, tensor de 7x7x256 a 14x14x128, con normalización por lote y activación ReLU
    X = Conv2DTranspose(filters     = 128, 
                        kernel_size = 3, 
                        strides     = 2, 
                        padding     = 'same')(X)
    X = BatchNormalization()(X)
    X = LeakyReLU(alpha = 0.01)(X)
    
    # Convolución transpuesta, tensor de 14x14x128, a 14x14x64 con normalización por lote y activación ReLU
    X = Conv2DTranspose(filters     = 64, 
                        kernel_size = 3, 
                        strides     = 1, 
                        padding     = 'same')(X)
    X = BatchNormalization()(X)
    X = LeakyReLU(alpha=0.01)(X)
    
    # Convolución transpuesta, tensor de 14x14x128 a 28x28x1, con activación tahn
    Y = Conv2DTranspose(filters     = 3, 
                        kernel_size = 3, 
                        strides     = 2, 
                        padding     = 'same',
                        activation  = 'tanh')(X)

    

    generator_model = Model(inputs = z, outputs = [Y], name ='generator')
    
    return generator_model
```

## Agregando etiquetas


Necesitamos que nuestro generador tenga también información de las etiquetas generadas de manera aleatoria, para ello, pedimos que este modelo reciba dos entradas
1. El lote de imágenes falsas.
2. Las etiquetas generadas.

Luego vamos a combinar esta información en una capa que codifique dichas etiquetas por categorías para luego pasarlas al resto de las capas. El modelo se ilustra a continuación

generador.jpg

```python
def build_generator(img_shape, z_dim, verbose=False):
    '''
    Genera una imagen de 28x28x1 a partir de un vector aleatorio de 100 entradas (espacio latente)
    '''
    # Batch de datos
    z           = Input(shape=(z_dim,))
    # batch de etiquetas
    image_class = Input(shape = (1,), dtype = np.float32)

    # 10 clases en MNIST
    cls = Flatten()(tf.keras.layers.Embedding(10, z_dim)(image_class))
    # hadamard product between z-space and a class conditional embedding
    h = tf.keras.layers.Multiply()([z, cls])

    # Pasa entrada unidimensional de dimensión 20 en un tensor de (7)(7)(256) tensor via un red Densa
    # luego la reformatea en un tensor de 7x7x128
    X = Dense(256 * 7 * 7, input_dim=z_dim) (h)
    X = Reshape((7, 7, 256))(X)

    # Convolución transpuesta, tensor de 7x7x256 a 14x14x128, con normalización por lote y activación ReLU
    X = Conv2DTranspose(filters     = 128, 
                        kernel_size = 3, 
                        strides     = 2, 
                        padding     = 'same')(X)
    X = BatchNormalization()(X)
    X = LeakyReLU(alpha = 0.01)(X)
    
    # Convolución transpuesta, tensor de 14x14x128, a 14x14x64 con normalización por lote y activación ReLU
    X = Conv2DTranspose(filters     = 64, 
                        kernel_size = 3, 
                        strides     = 1, 
                        padding     = 'same')(X)
    X = BatchNormalization()(X)
    X = LeakyReLU(alpha=0.01)(X)
    
    # Convolución transpuesta, tensor de 14x14x128 a 28x28x1, con activación tahn
    Y = Conv2DTranspose(filters     = 1, 
                        kernel_size = 3, 
                        strides     = 2, 
                        padding     = 'same',
                        activation  = 'tanh')(X)

    

    generator_model = Model(inputs = [z, image_class], outputs = [Y], name ='generator')
    
    return generator_model
```

Construye Generador

```python
generator = build_generator(img_shape, z_dim)
generator.summary()
```

# Discriminador

En discriminador es una red neuronal secuencial profunda que recibe de entrada un vector datos y produce una clasificación indicando si los datos son reales o falsos (fake).

discriminador.png

```python
def build_discriminator(img_shape, verbose=False):
    Xin  = Input(shape=(img_shape[0],img_shape[1],img_shape[2],))
    # Convolución2D tensor de 28x28x1 a 14x14x32 y activación Leaky ReLU
    X = Conv2D(filters     = 32, 
               kernel_size = 3, 
               strides     = 2, 
               input_shape = img_shape, 
               padding     = 'same')(Xin)
    #X = BatchNormalization()(X)
    X = LeakyReLU(alpha    = 0.01)(X)

    # Convolución2D tensor de 14x14x32 a 7x7x64, con normalización por lote y activación Leaky ReLU
    X = Conv2D(filters     = 64, 
               kernel_size = 3, 
               strides     = 2, 
               padding     = 'same')(X)
    X = BatchNormalization()(X)
    X = LeakyReLU(alpha    = 0.01)(X)
    
    # Convolución2D tensor de 7x7x64 a 3x3x128, con normalización por lote y activación Leaky ReLU
    X = Conv2D(filters     = 128, 
               kernel_size = 3, 
               strides     = 2,  
               padding     = 'same')(X)
    X = BatchNormalization()(X)
    X = LeakyReLU(alpha    = 0.01)(X)

    # Aplanado del tensor, y capa densa de salida de clasificacion con activación sigmoide
    X = Flatten()(X)

    fake = Dense(1, activation = 'linear', name = 'generator')(X)                    
    # aux = Dense(10, activation = 'softmax', name = 'auxiliary')(X)

    
    discriminator_model = Model(inputs = Xin, outputs = [fake], name ='discriminator')

    return discriminator_model
```

## Construye y compila el Discriminador


```python
# construye el discriminador 
discriminator = build_discriminator(img_shape)
discriminator.summary()
```

# Crítico Wasserstein

De manera similar al generador, necesitamos que el crítico conozca las etiquetas generadas, para ellos se las damos de manera similar a como se hizo en el generador

discriminador.jpg

```python
def build_Wasserstein_critic(img_shape, verbose = False):
    Xin = Input(shape=(img_shape[0],img_shape[1],img_shape[2],))
    '''# Agrego clase de vector
    image_class = Input(shape = (1,), dtype = np.float32)
    # 10 clases en MNIST
    cls = Flatten()(tf.keras.layers.Embedding(10, z_dim)(image_class))
    # hadamard product between z-space and a class conditional embedding
    h = tf.keras.layers.Multiply()([Xin, cls])'''

    # Convolución2D tensor de 28x28x1 a 14x14x32 y activación Leaky ReLU
    X = Conv2D(filters     = 32, 
               kernel_size = 3, 
               strides     = 2, 
               input_shape = img_shape, 
               padding     = 'same')(Xin)
    #X = BatchNormalization()(X)
    X = LeakyReLU(alpha    = 0.01)(X)

    # Convolución2D tensor de 14x14x32 a 7x7x64, con normalización por lote y activación Leaky ReLU
    X = Conv2D(filters     = 64, 
               kernel_size = 3, 
               strides     = 2, 
               padding     = 'same')(X)
    X = BatchNormalization()(X)
    X = LeakyReLU(alpha    = 0.01)(X)
    
    # Convolución2D tensor de 7x7x64 a 3x3x128, con normalización por lote y activación Leaky ReLU
    X = Conv2D(filters     = 128, 
               kernel_size = 3, 
               strides     = 2,  
               padding     = 'same')(X)
    X = BatchNormalization()(X)
    X = LeakyReLU(alpha    = 0.01)(X)

    # Aplanado del tensor, y capa densa de salida de clasificacion con activación sigmoide
    Y_out1 = Flatten()(X)
    Y_out2 = Dense(1, activation='sigmoid')(Y_out1)

    critic_model = Model(inputs = Xin, outputs = [Y_out2], name ='critic')

    return critic_model
```

```python
# construye el discriminador 
wasserstein_critic = build_Wasserstein_critic(img_shape)
wasserstein_critic.summary()
```

# Funciones de Pérdida

```python
# Funciones clásicas
cross_entropy = tf.keras.losses.BinaryCrossentropy(from_logits=True)
mae           = tf.keras.losses.MeanAbsoluteError(name='mean_absolute_error')
```

```python
Lambda = [0.5, 1.0]

def generator_loss_classic(fake_output) :
  return cross_entropy(tf.ones_like(fake_output), fake_output)

def generator_loss_v2(real_output, fake_output) :
  return - tf.keras.backend.log(tf.reduce_mean(fake_output))

def generator_loss(real_output, fake_output) :
  return tf.keras.backend.log(1.0 - tf.reduce_mean(fake_output))

def discriminator_loss_classic(real_output, fake_output) :
  # Pérdida de los verdaderos (1 vs y_hat)
  real_loss = cross_entropy(tf.ones_like(real_output), real_output)
  # Pérdida de los sintéticos (0 vs y_hat)
  fake_loss = cross_entropy(tf.zeros_like(real_output), real_output)
  # Suma de los dos tipos de errores
  total_loss = real_loss + fake_loss
  return total_loss

def discriminator_loss_v2 (real_output, fake_output) :
  # Pérdida de los verdaderos (1 vs y_hat)
  return tf.keras.backend.log(tf.reduce_mean(fake_output)) - tf.keras.backend.log(tf.reduce_mean(real_output))

def discriminator_loss(real_output, fake_output) :
  # Pérdida de los valores ( 1 vs y_hat)
  return - tf.keras.backend.log(tf.reduce_mean(real_output) * (1.0 - tf.reduce_mean(fake_output)))
```

## Implementación de la nueva función de pérdida

```python
def wasserstein_generator_loss (real_output, fake_output) :
  lossEM = tf.reduce_mean(tf.math.abs(tf.reduce_mean(real_output[0], 0) - tf.reduce_mean(fake_output[0], 0)))
  lossC  = tf.keras.backend.log(1.0 - tf.reduce_mean(fake_output[1]))
  return Lambda[0] * lossEM + Lambda[1] * lossC

def wasserstein_critic_loss(real_output, fake_output) :
  lossEM = tf.reduce_mean(tf.math.abs(tf.reduce_mean(real_output[0], 0) - tf.reduce_mean(fake_output[0], 0)))
  lossC  = - tf.keras.backend.log( tf.reduce_mean(real_output[1]) *  (1.0 - tf.reduce_mean(fake_output[1])))
  return Lambda[0] * lossEM + Lambda[1] * lossC
```

## Creamos modelos

```python
Losses = {'dcgan' : {'generator'     : generator_loss,
                     'discriminator' : discriminator_loss},
          'wgan'  : {'generator'     : wasserstein_generator_loss,
                     'discriminator' : wasserstein_critic_loss}
         }
Models = {'dcgan' : {'generator'     : generator,
                     'discriminator' : discriminator},
          'wgan'  : {'generator'     : generator,
                     'discriminator' : wasserstein_critic}
         }

# Para el paso de entrenamiento
modelname = 'wgan'
GAN_discriminator = Models[modelname]['discriminator']
GAN_generator     = Models[modelname]['generator']
# Pérdidas
Loss_discriminator = Losses[modelname]['discriminator']
Loss_generator     = Losses[modelname]['generator']
```

# Paso de entrenamiento

Definiremos un paso de entrenamiento de la GAN, que corresponde a hacer un paso de entrenamiento en cada modelo.

1. Carga lote de imágenes reales $ X $ y etiquetas $y$.

2. Genera un lote de variables latentes $ Z $

3. Genera un lote de variables sintéticas: $ \hat{X} = G(Z; \theta_g) $

4. Calcula las predicciones para $ \hat {Y} = D([X | \hat{X}], \theta_d) $

5. Calcula costos $ J^D $ y $ J^G$.

6. Calcula gradientes $\nabla_{\theta_d} J^D$ y $ \nabla_{\theta_g} J^G$ .

7. Actualiza pesos con un paso del algoritmo de optimización:

$$ \theta_d = \theta_d + \delta(\nabla_{\theta_d} J^D) $$ 
$$ \theta_g = \theta_g + \delta(\nabla_{\theta_g} J^G)$$ 

donde $\delta(\cdot)$ representa el paso del algoritmo utilizado.

```python
#El decorador `tf.function` indica que la función se "compila" para que pueda incluirse en un gráfo de calculo.
@tf.function
def train_step(images = None, labels = None, modelname = 'dcgan'):
    '''
    Implementa un paso de entrenamiento para la GAN
    
    Recibe como parámetros un semi-lote de imágenes reales
    '''

    # variables latentes (ruido Gaussiano), tantas como imagenes reales
    z = tf.random.normal([images.shape[0], z_dim])
    sampled_labels = np.random.randint(0, 10, images.shape[0])
    sampled_labels = sampled_labels.reshape((-1, 1))

    # Los siguientes pasos registaran (en "TAPE") para efectos de calcular gradientes
    # son dos "Tapes" para registrar los calculos de cada modelo
    with tf.GradientTape() as gen_tape, tf.GradientTape() as disc_tape:
        # genera el semi-lote de datos sintéticos
        # generated_images = GAN_generator([z,sampled_labels] , training=True)
        generated_images = GAN_generator(z, training=True)
        
        # calcula la predicción para datos verdaderos y falsos
        # real_output = GAN_discriminator([images,labels] ,  training = True)
        real_output = GAN_discriminator(images,  training = True)
        # fake_output = GAN_discriminator([generated_images,sampled_labels], training = True)
        fake_output = GAN_discriminator(generated_images, training = True)

        # calcula las pérdidas del Discriminador y del Generador
        disc_loss = Loss_discriminator(real_output, fake_output)
        gen_loss  = Loss_generator(real_output, fake_output)

    # Para cada modelo, calcula el gradiente de su función de costo respecto a sus pesos entrenables, 
    # haciendo retropropagación sobre los calculos realizados
    gradients_of_generator     = gen_tape.gradient(gen_loss, GAN_generator.trainable_variables)
    gradients_of_discriminator = disc_tape.gradient(disc_loss, GAN_discriminator.trainable_variables)

    # Hace el paso de actualizacion de los pesos
    generator_optimizer.apply_gradients    (zip(gradients_of_generator    , GAN_generator.trainable_variables))
    discriminator_optimizer.apply_gradients(zip(gradients_of_discriminator, GAN_discriminator.trainable_variables))
    
    return disc_loss, gen_loss
```

# Iteración del entrenamiento

```python
from PIL import Image
import imageio
import glob

# - - - - - - - - - - - - - - - - - - - - - - - - - - 
# Crea una collage de imágenes con el Generador y la guarda
# - - - - - - - - - - - - - - - - - - - - - - - - - - 

num_img_rows, num_img_cols = (4,4)
num_examples_to_generate = num_img_rows * num_img_cols

seed        = tf.random.normal([num_examples_to_generate, z_dim])
labels_seed = np.random.randint(0, 10, num_examples_to_generate)
labels_seed = labels_seed.reshape((-1, 1))

def generate_and_save_images(model, epoch, test_input, labels, pathdir=path_results):
    # Se pone entrenable en Falso porque esta en modo inferencia el 
    # predictions = model([test_input, labels], training=False)
    predictions = model(test_input, training=False)
    # predictions -> [0,255]
    predictions = predictions - tf.reduce_min(predictions)
    predictions = 255 * predictions / tf.reduce_max(predictions)
    data = tf.cast(predictions, tf.uint8)

    fig = plt.figure(figsize=(4,4))  

    for i in range(data.shape[0]):
        plt.subplot(4, 4, i+1)
        plt.imshow(data[i, :, :, :])
        plt.axis('off')

    plt.savefig(pathdir+'image_epch_{:04d}.png'.format(epoch))
    plt.show()
    
# - - - - - - - - - - - - - - - - - - - - - - - - - - 
# Despliega la imagen correspondiente a una época
# - - - - - - - - - - - - - - - - - - - - - - - - - - 

def display_image(epoch, pathdir=path_results, modelname = 'dcgan') :
    filename = '{}{}_image_epoch_{:04d}.png',format(pathdir, modelname, epoch)

    return Image.open(filename)

# ----------------------------------------------------
# Crea un gif animado con las imágenes de las épocas
# ----------------------------------------------------

def createGif(anim_file = 'gan.gif', pathdir = path_results, modelname = 'dcgan') :
  gifname = '{}{}_{}'.format(pathdir, modelname, anim_file)
  with imageio.get_writer(gifname, mode = 'I') as writer :
    finelames = glob.glob(pathdir + 'image*.png')
    filenames = sorted(finelames)
    for i, filename in enumerate(filenames) :
      image = imageio.imread(filename)
      if i % 5 == 0 :
        writer.append_data(image)

from tensorflow.keras.preprocessing.image import load_img, img_to_array

def saveGif(anim_file = 'gan.gif', pathdir = path_results, modelname = 'dcgan') :
  gifname   = '{}{}_{}'.format(pathdir, modelname, anim_file)
  filenames = glob.glob(pathdir + 'image*.png')
  filenames = sorted(filenames)

  print(gifname, filenames, len(filenames))
  
  images = []
  for i, filename in enumerate(filenames) :
      images.append(load_img(filename))
  images[1].save(gifname,
                 save_all = True,
                 append_images = images[2:],
                 optimize = False,
                 duration = 40,
                 loop = 0)
```

```python
from IPython import display
import time 
from tqdm import tqdm

def train(dataset, epochs, modelname = 'dcgan'):
    generator_losses     = [] 
    discriminator_losses = []
    for epoch in range(epochs):
        start = time.time()
        t = 0
        for image_batch, true_label in dataset:
            gen_loss, disc_loss = train_step(images = image_batch, labels = true_label, modelname = modelname)

        # solo registramos los costos en el último lote
        generator_losses.append(gen_loss)
        discriminator_losses.append(disc_loss)
        
        # Produce imágenes para crear el GIF
        if (epoch + 1) % each_save == 0:
            display.clear_output(wait = True)  # limpia el buffer
            generate_and_save_images(model = generator,
                                     epoch = epoch + 1,
                                     test_input = seed,
                                     labels = labels_seed,
                                     pathdir = path_results)
        if (epochs + 1) % save_epochs == 0 :
          checkpoint.save(file_prefix = checkpoint_prefix)
        print ('{} : Time for epoch {} is {} sec'.format(modelname, epoch + 1, time.time()-start))
    # Genera conjunto de imágenes para mostrar avances
    display.clear_output(wait = True)
    generate_and_save_images(model = generator,
                                     epoch = epoch + 1,
                                     test_input = seed,
                                     labels = labels_seed,
                                     pathdir = path_results)
    return np.array(generator_losses), np.array(discriminator_losses)
```
